```python
### Introduction
# Ce notebook présente les fonctions PyAutoGUI les plus utilisées.
# Une seule fonction par cellule, avec explication.

### Fonction moveTo()
# Déplace la souris à une position donnée (x, y) sur l'écran.
import pyautogui

pyautogui.moveTo(100, 100, duration=1)  # Déplace la souris en 1 seconde vers (100, 100)

### Fonction click()
# Clique à la position actuelle ou à une position précise si x et y sont donnés.
pyautogui.click(200, 200)

### Fonction moveRel()
# Déplace la souris par rapport à sa position actuelle.
pyautogui.moveRel(50, 0, duration=0.5)  # Déplace 50 pixels vers la droite

### Fonction dragTo()
# Glisse la souris jusqu'à une position donnée tout en maintenant le clic.
pyautogui.dragTo(300, 300, duration=1)

### Fonction dragRel()
# Glisse la souris par rapport à sa position actuelle.
pyautogui.dragRel(0, 50, duration=1)

### Fonction position()
# Renvoie la position actuelle de la souris.
pos = pyautogui.position()
print(pos)

### Fonction size()
# Renvoie la taille de l'écran (largeur, hauteur).
size = pyautogui.size()
print(size)

### Fonction screenshot()
# Prend une capture d'écran et la sauvegarde.
pyautogui.screenshot('capture.png')

### Fonction write()
# Tape une chaîne de caractères comme si on la tapait au clavier.
pyautogui.write('Hello world!', interval=0.1)

### Fonction press()
# Appuie sur une touche du clavier.
pyautogui.press('enter')

### Fonction hotkey()
# Combine plusieurs touches comme un raccourci clavier.
pyautogui.hotkey('ctrl', 'c')

### Fonction alert()
# Affiche une boîte de dialogue d'alerte simple.
pyautogui.alert('Fin du test PyAutoGUI !')
```
